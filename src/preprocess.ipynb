```python
import pandas as pd
import numpy as np
from math import comb as comb
```

```python
df_original = pd.read_csv('../data/Wimbledon_featured_matches.csv')
```

### Modify

```python
df_original['match_id'] = df_original['match_id'].rank(method='dense').astype(int)
```

```python
cur_set_id = 1
last_set_no = 1
df_original['set_id'] = 0
for i, row in df_original.iterrows():
    df_original.loc[i, 'set_id'] = cur_set_id
    if row['set_no'] != last_set_no:
        cur_set_id += 1
        last_set_no = row['set_no']
```

```python
cur_game_id = 1
last_game_no = 1
df_original['game_id'] = 0
for i, row in df_original.iterrows():
    df_original.loc[i, 'game_id'] = cur_game_id
    if row['game_no'] != last_game_no:
        cur_game_id += 1
        last_game_no = row['game_no']
```

```python
score_mapping = {'0': 0, '15': 1, '30': 2, '40': 3, 'AD': 4}
df_original["p1_score"] = df_original["p1_score"].map(lambda x: score_mapping[x] if x in score_mapping else x).astype(int)
df_original["p2_score"] = df_original["p2_score"].map(lambda x: score_mapping[x] if x in score_mapping else x).astype(int)
```

```python
winner_shot_type_mapping = {'F': 0, 'B': 1}
df_original["winner_shot_type"] = df_original["winner_shot_type"].map(lambda x: winner_shot_type_mapping[x] if x in winner_shot_type_mapping else x).astype(int)
```

```python
# fill na with average speed
df_original["speed_mph"].fillna(df_original["speed_mph"].mean(), inplace=True)
```

```python
# B / 0: Body, BC / 1: Body/Center, BW / 2: Body/Wide, C / 3: Center, W / 4: Wide"
serve_width_mapping = {'B': 0, 'BC': 1, 'BW': 2, 'C': 3, 'W': 4}
df_original["serve_width"] = df_original["serve_width"].fillna(0).map(lambda x: serve_width_mapping[x] if x in serve_width_mapping else x).astype(int)
```

```python
# "CTL / 0: Close To Line, NCTL / 1: Not Close To Line"
serve_depth_mapping = {'CTL': 0, 'NCTL': 1}
df_original["serve_depth"] = df_original["serve_depth"].fillna(0).map(lambda x: serve_depth_mapping[x] if x in serve_depth_mapping else x).astype(int)
```

```python
# "D / 0: Deep, ND / 1: Not Deep"
return_depth_mapping = {'D': 0, 'ND': 1}
df_original["return_depth"] = df_original["return_depth"].fillna(0).map(lambda x: return_depth_mapping[x] if x in return_depth_mapping else x).astype(int)
```

```python
df_original.to_csv('../data/Wimbledon_featured_matches_modified.csv', index=False)
```

## Win Rate

```python
# hyper
set_win_rate = 0.5
server_game_win_rate = 0.8
receiver_game_win_rate = 1 - server_game_win_rate
server_score_win_rate = 0.8
receiver_score_win_rate = 1 - server_score_win_rate
```

```python
def win_rate_by_set(p1_sets, p2_sets): # return p1_win_rate
    p1_win_rate = 0
    m = 3 - p1_sets
    n = 5 - p1_sets - p2_sets
    for i in range(m, n + 1):
        p1_win_rate += comb(n, i)*set_win_rate**i*(1-set_win_rate)**(n-i)
    return p1_win_rate
```

```python
def win_rate_by_game(server, # 1 or 2 first server
                     p1_games, 
                     p2_games
                     ):
    if(server == 1):
        p1_win_rate_single = server_game_win_rate
        p2_win_rate_single = receiver_game_win_rate
    else:
        p1_win_rate_single = receiver_game_win_rate
        p2_win_rate_single = server_game_win_rate
    if(p1_games >= 5 and p2_games >= 5): # After 5:5(including), tie break
        p1_games -= 5
        p2_games -= 5
        p1_win_rate = 0
        m = 2 - p1_games
        n = 3 - p1_games - p2_games
        for i in range(m, n + 1):
            p1_win_rate += comb(n, i)*p1_win_rate_single**i*p2_win_rate_single**(n-i)
        return p1_win_rate
    elif(p1_games == 6 and p2_games < 5):
        return 1
    elif(p2_games == 6 and p1_games < 5):
        return 0
    else: # Before 5:5, 
        tie_break_prob = \
            comb(10 - p1_games - p2_games, 5 - p1_games)\
            *p1_win_rate_single**(5 - p1_games)\
            *p2_win_rate_single**(5 - p2_games)
        p1_win_rate_with_tie_break = win_rate_by_game(server, 5, 5)
        if(p2_games >= 5):
            p1_win_rate_without_tie_break = 0
        else:
            m = 5 - p1_games
            n = 9 - p1_games - p2_games
            p1_win_rate_without_tie_break = 0
            for i in range(m, n + 1):
                p1_win_rate_without_tie_break += comb(n, i)*p1_win_rate_single**i*p2_win_rate_single**(n-i)
                
        return p1_win_rate_with_tie_break*tie_break_prob + p1_win_rate_without_tie_break*(1 - tie_break_prob)
```

```python
def win_rate_by_score(server, # 1 or 2 first server
                      p1_score, 
                      p2_score, 
                     ):
    if(server == 1):
        p1_win_rate_single = server_score_win_rate
        p2_win_rate_single = receiver_score_win_rate
    else:
        p1_win_rate_single = receiver_score_win_rate
        p2_win_rate_single = server_score_win_rate
    if(p1_score >= 3 and p2_score >= 3): # After 3:3(including)
        min_score = min(p1_score, p2_score)
        p1_score -= min_score
        p2_score -= min_score
        if(p1_score == 1): # p1 - AD
            return p1_win_rate_single + (1 - p1_win_rate_single)*win_rate_by_score(server, 3, 3)
        elif(p2_score == 1): # p2 - AD
            return p1_win_rate_single*win_rate_by_score(server, 3, 3)
        else: # score all
            p1_win_rate = 0
            p1_single = p1_win_rate_single
            # x = p1_single**2 + 2*p1_single*(1 - p1_single)*x
            x = (p1_single)**2 / (1 - 2*p1_single*(1 - p1_single))
            return x
    elif(p1_score == 4 and p2_score < 3):
        return 1
    elif(p2_score == 4 and p1_score < 3):
        return 0
    else: # Before 3:3
        tie_break_prob = \
            comb(6 - p1_score - p2_score, 3 - p1_score)\
            *p1_win_rate_single**(3 - p1_score)\
            *p2_win_rate_single**(3 - p2_score)
        p1_win_rate_with_tie_break = win_rate_by_score(server, 3, 3)
        if(p2_score >= 3):
            p1_win_rate_without_tie_break = 0
        else:
            m = 4 - p1_score
            n = 6 - p1_score - p2_score
            p1_win_rate_without_tie_break = 0
            for i in range(m, n + 1):
                p1_win_rate_without_tie_break += comb(n, i)*p1_win_rate_single**i*p2_win_rate_single**(n-i)
        return p1_win_rate_with_tie_break*tie_break_prob + p1_win_rate_without_tie_break*(1 - tie_break_prob)
        # return p1_win_rate_without_tie_break
```
